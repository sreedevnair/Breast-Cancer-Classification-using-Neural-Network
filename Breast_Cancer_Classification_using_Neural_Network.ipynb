```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print("Breast Cancer Data :-")
print(breast_cancer_dataset.data)

print("\nBreast Cancer Target Value :-")
print(breast_cancer_dataset.target)

print("\nBreast Cancer Columns :-")
breast_cancer_dataset.feature_names
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
# Adding target column to our dataframe
df['label'] = breast_cancer_dataset.target
```

```python
df.describe()
```

```python
df.label.value_counts()

# 1 -> Benign (Non Cancer)
# 0 -> Malignant (Cancer)
```

```python
df.groupby('label').mean()
```

Splitting Our Dataset

```python
X = df.drop(columns='label')
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

Standardizing Our Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)
```

**Building the Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation="relu"),
    keras.layers.Dense(2, activation="sigmoid")
])
```

```python
model.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
history = model.fit(X_train_std, y_train, validation_split=0.1, epochs=10)
```

```python
# Visualizing Accuracy and Loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

plt.legend(['Training', 'Validation'], loc = 'lower right')
```

```python
# Visualizing Accuracy and Loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.legend(['Training', 'Validation'], loc = 'lower right')
```

**Finding accuracy and loss on the test data**

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
```

```python
print("Accuracy of the model is :", accuracy)
print("Loss of the model is :", loss)
```

```python
y_pred = model.predict(X_test_std)
```

```python
print(y_pred.shape)
print(y_pred[0])
```
